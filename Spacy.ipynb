```python
import numpy as np
import pandas as pd
import spacy
# Import the english language model
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv("fake.csv")
df.head()
```

```python
df.shape
```

```python
sample_df = df.sample(frac=0.2, replace=True, random_state=1) #20% sample size
```

```python
sample_df['text'].head
```

```python
sample_df.text = sample_df.text.astype(str)
```

```python
%%time
def nameDetection(txt):
    doc = nlp(txt)
    persons = [ent.text for ent in doc.ents if ent.label_ == 'PERSON']
    #print(f'persons is {persons}')
    return persons

#df["text"] = df["text"].apply(lambda x:nameDetection(x['text'],axis=1) #row by row
namesExist = []
for index, row in sample_df.iterrows():
    namesExist.extend(nameDetection(row['text']))

    
```

```python
%%time
#Remove Duplicates in list

namesExist = list(dict.fromkeys(namesExist))

print(namesExist)
```
